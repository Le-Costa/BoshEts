Crie um array de 2 linhas e 3 colunas e faça as seguintes operações:
Faça um reshape de 3 linhas e 2 colunas;
Some a primeira linha por 5;
Multiplique a segunda linha por 3;
E divida a segunda coluna por 2.

```python
matriz = np.random.randint(0,10,6)
n = np.array(matriz).reshape((2,3))
i = n.reshape(3,2)
print(i)
print("Soma:")
soma = i[0]+5
print(soma)
multiplicacao = i[1]*3
print("Multi",multiplicacao)
i[:,1] = i[:,1]/2

```

Crie uma função que recebe um número inteiro n e retorna um array n x n do NumPy , com as seguintes regras:
Os elementos da diagonal principal devem valer 1
Os elementos da diagonal secundária devem valer 2
Os demais elementos devem valer 0
Quando um elemento pertence às duas diagonais (isso só ocorre se n for ímpar), ele deve valer 3

```python
n = int(input("Insira o tamanho da matriz: "))
l=0
c=0
matriz= np.zeros((n,n))
for i in range(n):
    matriz[l][c]=1
    l+=1
    c+=1


t = len(matriz[0])
l=0
c=t-1
for i in range(t):
    matriz[l][c]=2
    l+=1
    c-=1
print(matriz)
```

```python
n = int(input("Insira o tamanho da matriz: "))
matriz= np.zeros((n,n))
for i in range(n):
    matriz[i,i]=1
    matriz[i,n-1-i]+=2

print(matriz)
```

```python
n = 3
matriz = np.random.randint(0,10,(n,n))
m= matriz.reshape((3,3)) #coloca em formato de tamanho
print(m,"-")
nova = np.zeros((n,n)) #nova matriz com zeros

for i in range(n):
    for j in range(n):
        if i != 0: 
            nova[i,j] +=m[i-1,j]#some linha de baixo
        if j!=0:
            nova[i,j] += m[i,j-1]#some coluna de tras
        if i != n-1:
            nova[i,j]+= m[i+1,j] #some linha de baixo
        if j != n-1:
            nova[i,j]+= m[i,j+1] # some coluna de cima

print(nova)
          


```
