```python
import pandas as pd
titanic= pd.read_csv("titanic.csv",sep=",") 
```

```python
#criar um afunção e aplicar no dataframe
def soma_sibsp_parch(linha):
    return linha["SibSp"] + linha["Parch"]
```

```python
nova_coluna = titanic.apply(soma_sibsp_parch, axis=1)
print(nova_coluna)

titanic["Relatives"]= nova_coluna #criação de nova coluna
```

criar coluna de faixa etaria:  12 - criança , >= 12 & < 18 adolescente > 18 adulto

```python

#dt= titanic
media= int(titanic["Age"].mean())
titanic["Age"]=round(titanic["Age"].fillna(titanic["Age"].mean()),0)


```

```python
def FaixaEtaria(linha):
    if linha["Age"] < 12.0:
        return "Criança"
    elif (linha["Age"]>= 12.0 and linha["Age"] <18.0):
        return "Adolescente"
    elif (linha["Age"]>=18.0 and linha["Age"] <65.0):
        return "Adulto"
    elif linha ["Age"]>=65:
        return "Idoso"
    else:
        return "Indefinido"

New = titanic.apply(FaixaEtaria, axis =1)
#titanic[titanic["FaixaEtaria"]=="Idoso"]
titanic["FaixaEtaria"] = New
titanic.head(10)


```

```python
agrupado = titanic.groupby("FaixaEtaria").count()


def calculo(linha):
    return round((linha["Pclass"]/agrupado["Survived"].sum()*100),2)

percentual = agrupado.apply(calculo,axis=1)

agrupado["Percentual"]= percentual 
print(agrupado[["PassengerId","Percentual"]])
```

```python
def CalculoPerc(val,total):
    percent = (val/total)*100
    return f"{percent:.2f}"

```
