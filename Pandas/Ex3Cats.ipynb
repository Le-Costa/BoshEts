```python

#A razão entre o peso (Kg) e comprimento (m) dos gatos com mais de 6 meses de idade.
def razao (linha):
    if linha["Age_in_months"]>6:
        return linha["Weight"]/(linha["Body_length"]/100)

cats["Razao peso*comprimento"]= cats.apply(razao,axis=1)
cats.head()
```

```python
#A porcentagem do dia que cada gato Ragdoll passa dormindo (como string).
def horas(linha):
    if linha["Breed"] =="Ragdoll":
        return str(round((linha["Sleep_time_hours"]/24)*100,2))

Ragdoll = cats[cats["Breed"]=="Ragdoll"]
Ragdoll["Percentual de sono"] = Ragdoll.apply(horas,axis=1)

Ragdoll[["Breed","Sleep_time_hours","Percentual de sono"]]
```
