Faça o resumo estatístico por gênero (coluna Sex), contendo a média de idade, o número de sobreviventes e o número de vítimas. (Utilize groupby)

```python
quadro_etario = titanic.groupby("Faixa_Etaria")["Faixa_Etaria"].count().to_frame(name="Quantidade")
quadro_etario['Porcentagem'] = round((quadro_etario['Quantidade'] / quadro_etario['Quantidade'].sum()) * 100,2)
print(quadro_etario) 
```
