```python
import pandas as pd
tp = pd.read_csv ("tips.csv", sep=";")


```

```python

#Crie uma coluna tip_pct com a porcentagem de gorjeta em relação ao total da conta.
def percentual(linha):
    return round((linha["tip"]/ linha["total_bill"])*100)

tp["PercentualTip"] = tp.apply(percentual,axis=1)
tp.head()
```

```python
#Encontre a mesa com maior gorjeta percentual.
tp = tp.sort_values(by="PercentualTip",ascending=False)
tp.head(1)
```
